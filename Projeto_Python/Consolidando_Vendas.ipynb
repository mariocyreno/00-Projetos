```python
import pandas as pd
import os
```

```python
# Consoliudando os arquivos utilizando o concat
diretorio = r'C:\Users\Administrador\Documents\00-Projetos\Projeto_Python\Origem\arquivos_excel\Consolidar'
dados = []
for arquivo in os.listdir(diretorio):
    if arquivo.endswith('.xlsx'):
        caminho_arquivo = os.path.join(diretorio, arquivo)
        df = pd.read_excel(caminho_arquivo)
        dados.append(df)
consolidado = pd.concat(dados)
```

```python
# Salvando o arquivo consolidado
pasta_destino = r'C:\\Users\\Administrador\\Documents\\00-Projetos\\Projeto_Python\\Origem\\arquivos_excel\\Consolidar\\Arquivos consolidados\\'
nome_arquivo = 'Vendas_Consolidadas.xlsx'
caminho_completo = pasta_destino + nome_arquivo
if os.path.exists(caminho_completo):
    os.remove(caminho_completo)
consolidado.to_excel(caminho_completo, index = False)
```
