```python
import pandas as pd
import os                       # >> importando a biblioteca para manipulação de arquivos e diretórios
from dotenv import load_dotenv  # >> importando a função para carregar as variáveis de ambiente do arquivo .env

caminho_do_arquivo_env = r"C:\Users\Administrador\Documents\00-Projetos\Projeto_Python\Caminhos\caminho.env" #Caminho Completo
load_dotenv(dotenv_path=caminho_do_arquivo_env)
df = pd.read_csv(os.getenv("CLIENTE_CSV"), delimiter = ',')

df['email'] = df['email'].fillna('Não Cadastrado')
df['cell_phone'] = df['cell_phone'].fillna('Não Informado')
df['state'] = df['state'].fillna('Não Informado')
df['street'] = df['street'].fillna('Não Informado')
df['additionals'] = df['additionals'].fillna('Sem Informações Adicionais')
df['number'] = df['number'].fillna('Sem Número')
df['country'] = 'Br'

df['full_name'] = df['first_name'] + ' ' + df['last_name']

df['created_at'] = pd.to_datetime(df['created_at'])
df['purchased_at'] = pd.to_datetime(df['created_at']).dt.strftime('%H:%M:%S')
#df['created_at'] = df['created_at'].dt.strftime('%d/%m/%y %H:%M:%S')
df['created_at'] = df['created_at'].dt.strftime('%d/%m/%y')

colunas = ['created_at',
           'purchased_at',
           'full_name',
           'email',
           'cell_phone',
           'street',
           'state',
           'number',
           'additionals',
           'country']
                         
dfcolunas = df[colunas]
dfcolunas
```
